```python
import pandas as pd
import numpy as np
from wordcloud import WordCloud
import matplotlib.pyplot as plt
from nltk.corpus import stopwords
from nltk.tokenize import word_tokenize
import nltk

nltk.download('punkt')
nltk.download('stopwords')
```

```python
#data set
documents_df = pd.read_csv(r'C:\Users\ftsms\Downloads\docsdata.csv', encoding='utf-8')
documents = documents_df['text'].tolist()
combined_text = ' '.join(documents)
```

```python
stop_words = set(stopwords.words('english'))
tokens = [word for doc in documents for word in word_tokenize(doc.lower())
          if word.isalpha() and word not in stop_words]
text_for_wordcloud = ' '.join(tokens)
```

```python
# search query
query = "information"
query = query.lower()
frequencies = [word_tokenize(doc.lower()).count(query) for doc in documents]
```

```python
#scatter plot
colors = np.random.rand(20)

plt.figure(figsize=(7, 5))
plt.scatter(range(len(frequencies)), frequencies, c=colors, alpha=0.5)
plt.title(f"Frequency of '{query}' in Documents")
plt.xlabel('Document ID')
plt.ylabel('Frequency of Query Term')
plt.show()
```

```python
#wordcloud 
wordcloud = WordCloud(width=800,
                      height=400,
                      background_color='white').generate(text_for_wordcloud)
plt.figure(figsize=(8, 8))
plt.imshow(wordcloud, interpolation='bilinear')
plt.axis('off')  # Remove axis labels
plt.show()
```
